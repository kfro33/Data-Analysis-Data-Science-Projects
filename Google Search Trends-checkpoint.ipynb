```python
import pandas as pd
import matplotlib.pyplot as plt
from pytrends.request import TrendReq #Google Trends API for analysis, must do pip install pytrends
trends = TrendReq()
```

<h1 style= text-align:center>Top 15 Countries which searched on Google "Pizza Hut"</h1>

```python
trends.build_payload(kw_list=["Pizza Hut"])
data = trends.interest_by_region()
data = data.sort_values(by="Pizza Hut", ascending=False)
data = data.head(15)
print(data)
```

```python
data.reset_index().plot(x="geoName", y="Pizza Hut", figsize=(20,15), kind="bar")
plt.style.use('grayscale')
plt.show()
```

<h1 style= text-align:center>Taking a look into the Google Search Trend History for "Pizza Hut"</h1>

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=["Pizza Hut"])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(20,15))
data["Pizza Hut"].plot()
plt.style.use('grayscale')
plt.title('Total Google Searches for Pizza Hut', fontweight="bold")
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
