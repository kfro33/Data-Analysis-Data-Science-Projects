```python
import pandas as pd
import matplotlib.pyplot as plt
from pytrends.request import TrendReq #Google Trends API for analysis, must do pip install pytrends
trends = TrendReq()
```

<h1 style=text-align:center>Top 15 Countries which searched on Google "Data Science"</h1>

```python
trends.build_payload(kw_list=["Data Science"])
data = trends.interest_by_region()
data = data.sort_values(by="Data Science", ascending=False)
data = data.head(15)
print(data)
```

```python
data.reset_index().plot(x="geoName", y="Data Science", figsize=(20,15), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

<h1 style=text-align:center>Taking a look into the Google Search Trend History for "Data Science"</h1>

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=["Data Science"])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(20,15))
data["Data Science"].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Data Science', fontweight="bold")
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
